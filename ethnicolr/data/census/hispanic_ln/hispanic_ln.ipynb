### Parse Hispanic Last Name Census Data

```python
import pandas as pd
```

```python
df = pd.read_csv("26 Appendix E Census List of Spanish Surnames.csv")
df.tail(50)
```

```python
print(df.shape)
## Remove rows with Nans
sdf = df[df.isna().sum(axis = 1) < 3]
print(sdf.shape)
sdf.tail()
```

```python
ldf = pd.melt(sdf)
tdf = ldf[["value"]]
fin_df = tdf[~(tdf['value'].isna())]
```

```python
fin_df.value_counts()
```

```python
fin_df = fin_df[~fin_df.value.isin(["EZ", "Z", "O"])]
fin_df.rename(columns={"value": "hispanic_last_name"})
```

```python
fin_df.to_csv("census_hispanic_ln.csv", index = False)
```
